```python
def create_sizes_list(data, size_rev_idx):
    size_list = []
    for car in data:
        s = size_rev_idx[car['Vehicle Size']]
        size_list.append((car['Year_Make_Model'], s, car['MSRP']))
    
    return size_list
    
```
